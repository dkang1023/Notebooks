## INTERMEDIATE LEVEL

```python
# count how many movies have each of the content ratings
movies.content_rating.value_counts(dropna=False)
```

```python
# use a visualization to display that same data, including a title and x and y labels
movies.content_rating.value_counts().plot(kind='bar', title= 'Bar Plot of Movie Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
```

```python
# convert the following content ratings to "UNRATED": NOT RATED, APPROVED, PASSED, GP
movies.content_rating.replace(to_replace={'NOT RATED', 'APPROVED', 'PASSED', 'GP'}, value='UNRATED', regex=True, inplace=True)
movies.content_rating.value_counts()
# Stack overflow code cited https://stackoverflow.com/questions/36072626/pandas-replace-multiple-values-at-once
```

```python
# convert the following content ratings to "NC-17": X, TV-MA
movies.content_rating.replace(to_replace={'X', 'TV-MA'}, value='NC-17', regex=True, inplace=True)
movies.content_rating.value_counts()
```

```python
# count the number of missing values in each column
movies.isnull().sum()
```

```python
# if there are missing values: examine them, then fill them in with "reasonable" values
movies[movies.content_rating.isnull()]
movies.content_rating.fillna(value='NOT RATED', inplace=True)

# drinks.continent.fillna(value='NA', inplace=True)
```

```python
# calculate the average star rating for movies 2 hours or longer,
# and compare that with the average star rating for movies shorter than 2 hours
movies[movies.duration >= 120].star_rating.describe()
```

```python
movies[movies.duration < 120].star_rating.describe()
# Movies that are longer than 2 hours have a higher mean star rating than those shorter.  
```
